### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
for i, fila in tienda.iterrows():
  precio = fila["Precio"]
  costoEnvio = fila["Costo de envío"]
  cantidad = fila["Cantidad de cuotas"]

  precio = precio + costoEnvio
  precio = precio * cantidad #Aqui ya esta c/ venta con su envio

  print (precio) #aqui ya tenemos todos los precios reales actualizados por cada producto y compra con respectivo envio.
  #Esto arroja una lista de todo individualmente
  type(precio)


```

```python
tienda["Precio Total"] = (tienda["Precio"] + tienda["Costo de envío"]) * tienda["Cantidad de cuotas"]
total_ventas_tienda = tienda["Precio Total"].sum()

print("Total de ventas de esta tienda:", total_ventas_tienda)
```

```python
total_ventas_1 = 0

for i, fila in tienda.iterrows():
    precio = fila["Precio"]
    costoEnvio = fila["Costo de envío"]
    cantidad = fila["Cantidad de cuotas"]

    precio_total = (precio + costoEnvio) * cantidad
    total_ventas_1 += precio_total

# Aquí ya esta el total de ventas de esta tienda
print("Total de ventas de tienda 1:", total_ventas_1)
```

```python
total_ventas_2 = 0

for i, fila in tienda2.iterrows():
    precio = fila["Precio"]
    costoEnvio = fila["Costo de envío"]
    cantidad = fila["Cantidad de cuotas"]

    precio_total = (precio + costoEnvio) * cantidad
    total_ventas_2 += precio_total

# Aquí ya esta el total de ventas de esta tienda
print("Total de ventas de tienda 2:", total_ventas_2)
```

```python
total_ventas_3 = 0

for i, fila in tienda3.iterrows():
    precio = fila["Precio"]
    costoEnvio = fila["Costo de envío"]
    cantidad = fila["Cantidad de cuotas"]

    precio_total = (precio + costoEnvio) * cantidad
    total_ventas_3 += precio_total

# Aquí ya esta el total de ventas de esta tienda
print("Total de ventas de tienda 3:", total_ventas_3)
```

```python
total_ventas_4 = 0

for i, fila in tienda4.iterrows():
    precio = fila["Precio"]
    costoEnvio = fila["Costo de envío"]
    cantidad = fila["Cantidad de cuotas"]

    precio_total = (precio + costoEnvio) * cantidad
    total_ventas_4 += precio_total

# Aquí ya esta el total de ventas de esta tienda
print("Total de ventas de tienda 4:", total_ventas_4)
```

```python
# Paso 1: Asegúrate de tener la columna 'Cantidad de cuotas'
# Paso 2: Multiplica la calificación por la cantidad
tienda["Ponderado"] = tienda["Calificación"] * tienda["Cantidad de cuotas"]
promedio_ponderado1 = 0

# Paso 3: Suma ponderada dividido entre total de cuotas
total_ponderado1 = tienda["Ponderado"].sum()
total_cuotas = tienda["Cantidad de cuotas"].sum()

promedio_ponderado1 = total_ponderado1 / total_cuotas

print("Calificación promedio ponderada 1:", promedio_ponderado1)
```
